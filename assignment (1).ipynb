# reviewing duplicated data in a column

```python
# reviwing the duplicate data in CGPA

data["CGPA"].duplicated().sum()      
```

# removing the duplicated data 

```python
# if there is any duplicated data , remove data using drop_duplicates command for individual columns

data=data.drop_duplicates(("CGPA"))
```

```python
# reviewing the data after removing the duplicated data

data["CGPA"].duplicated().sum()  
```

# replacing null values 

```python
# using numpy library replacing null values with "hi"

data.replace(np.nan,"hi")
```

```python
data.head()
```

# the mean of "Fanancial Stress"

```python
data["Financial Stress"].mean()
```

# filling the null values using "bfill"

```python
# replacing the nul values using bfill (backword filling method). In which previous value is assigned to next null values

data.fillna(method="bfill")
```

```python
data.head()   # reviewing the data 
```

```python
data.isnull().sum()     # reviweing null values
```

# refilling null values using ffill(forward filing method)

```python
# using ffill command to replace null values . In this method forward values is assigned to previous null value 

data.fillna(method="ffill")
```
